```python
data1 = data.rename({'Unnamed: 0':'slno','User.ID':'userid','Book.Title':'title','Book.Rating':'rating'},axis=1)
data1
```

```python
data2['rating'].value_counts()
```

```python
data3 = data2.pivot_table(index='userid',
                     columns='title',
                     values='rating')
data3
```

```python
#impute those Nans with 0 values
data3.fillna(0, inplace=True)
data3
```

```python
#Calculating Cosine similarities between users
from sklearn.metrics import pairwise_distances
from scipy.spatial.distance import cosine, correlation
```

```python
data4 = 1 - pairwise_distances(data3.values,metric='cosine')
data4
```

```python
#store the results in dataframe
data5 = pd.DataFrame(data4)
data5
```

```python
#Set the index and column names to user ids 
data5.index = data2.userid.unique()
data5.columns = data2.userid.unique()
```
