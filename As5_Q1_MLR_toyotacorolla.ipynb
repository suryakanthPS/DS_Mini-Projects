## Model building

```python
import statsmodels.formula.api as smf
model=smf.ols('Price~Age+KM+HP+cc+Doors+Gears+QT+Weight',data=data3).fit()
```

## Model testing

```python
#finding coefficient parameter
model.params
#y=mx+c ; years_exp=m ; intercept=c
```

```python
#finding t values
model.tvalues
```

```python
model.pvalues
```

```python
np.round(model.pvalues,6)
```

```python
#R squared values
(model.rsquared,model.rsquared_adj) #model accuracy is 86.1%
```

## Model prediction

```python
#Manual prediction of price based on multiple parameters
new_data=pd.DataFrame({'Age':12,"KM":40000,"HP":80,"cc":1300,"Doors":4,"Gears":5,"QT":69,"Weight":1012},index=[0])
new_data
```
