```python
#import libraries
import pandas as pd
import numpy as np
```

```python
#import data
data = pd.read_csv("/content/Fraud_check.csv")
data
```

## EDA

```python
data.info()
```

```python
data.describe()
```

```python
data1 =data.copy()
```

```python
data1.insert(6,'tax_class'," ")

for i in range(0,len(data1['Taxable.Income']),1):
    if data1['Taxable.Income'].iloc[i]<=30000:
        data1['tax_class'].iloc[i]='Risky'
    else:
        data1['tax_class'].iloc[i]='Good'
data1
```

```python
from sklearn import preprocessing
le = preprocessing.LabelEncoder()
data1['Undergrad'] = le.fit_transform(data1['Undergrad'])
data1['Urban'] = le.fit_transform(data1['Urban'])
data1['Marital.Status'] = le.fit_transform(data1['Marital.Status'])
data1['tax_class'] = le.fit_transform(data1['tax_class'])
data1
```

```python
data1.tax_class.value_counts()
```

```python
data1 = data1.drop(['Taxable.Income'], axis=1)
```

## Model Building

```python
#Dividing the data into independent and dependent variable
X = data1.drop(['tax_class'],axis=1)
y = data1['tax_class']
```

```python
#Splitting the dataset into train and test dataset
from sklearn.model_selection import train_test_split
X_train, X_test ,y_train, y_test = train_test_split(X,y, test_size=0.2, random_state=40)
```

### Model building by classifier method - categorical variable

```python
#fitting Random forest classification to the training set
from sklearn.ensemble import RandomForestClassifier
classifier = RandomForestClassifier(n_estimators=10, max_depth = 3)
classifier.fit(X_train,y_train)
```

```python
#Model prediction
y_pred = classifier.predict(X_test)
y_pred
```

```python
y_test.values
```

```python
#Model accuracy
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_pred, y_test)
print(cm)

from sklearn.metrics import accuracy_score
acc = accuracy_score(y_pred, y_test)
print("Accuracy=", acc*100)
```

```python
#fitting Random forest classification to the training set
from sklearn.ensemble import RandomForestRegressor
regressor = RandomForestRegressor(n_estimators=3, max_depth = 3)
regressor.fit(X_train,y_train)
```

```python
y.pred = regressor.predict(X_test)
y_pred
```

```python
#Model accuracy
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_pred, y_test)
print(cm)

from sklearn.metrics import accuracy_score
acc = accuracy_score(y_pred, y_test)
print("Accuracy=", acc*100)
```
