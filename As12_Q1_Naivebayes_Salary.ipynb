```python
#import libraries
import pandas as pd
import numpy as np
```

```python
#import data
data_train = pd.read_csv("/content/SalaryData_Train.csv")
data_train
```

```python
data_test = pd.read_csv ("/content/SalaryData_Test.csv")
data_test
```

```python
#one hot encoding for categorical variables
data_train_encode = pd.get_dummies(data_train,columns=['workclass', 'education', 'maritalstatus', 'occupation', 'relationship', 
                                 'race', 'sex', 'native'])
data_train_encode
```

```python
data_train_encode.info()
```

## Model building

```python
#Dividing the data into independent and dependent variable
X = data_train_encode.drop(['Salary'], axis=1)
y = data_train_encode['Salary']
```

```python
X.info()
```

```python
#Splitting the dataset into train and test dataset
from sklearn.model_selection import train_test_split
X_train, X_test ,y_train, y_test = train_test_split(X,y, test_size=0.3, random_state=0)
```

```python
#Feature scaling
from sklearn.preprocessing import StandardScaler
sc_X=StandardScaler()
X_train = sc_X.fit_transform(X_train)
X_test = sc_X.transform(X_test)
```

```python
X_train.shape
```

```python
#Fitting naive bayes model
from sklearn.naive_bayes import GaussianNB
classifier = GaussianNB()
classifier.fit(X_train,y_train)
```

## Model prediction

```python
y_pred = classifier.predict(X_test)
y_pred
```

```python
y_test
```

## Model Accuracy

```python
#Confusion matrix
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test,y_pred)
cm
```

```python
#accuracy score
from sklearn.metrics import accuracy_score
accuracy_score = accuracy_score(y_pred,y_test)
accuracy_score*100
```
